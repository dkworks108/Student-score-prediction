```python
import cv2
```

```python
image = cv2.imread("bird.jpg")

```

```python
(h,w) = image.shape[:2]

center = (w//2 , h//2)

M = cv2.getRotationMatrix2D(center, 90, 2.0)

rotate = cv2.warpAffine(image, M, (w , h))

cv2.imshow("rotated image" , rotate)
cv2.imshow(" Orignal image", image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
new_img =  cv2.imread("bird.jpg")

(w_, h_)  = new_img.shape[:2] 

cent = (w_//2, h_//2)

M = cv2.getRotationMatrix2D(cent, 45 , 1.0)

rot = cv2.warpAffine(new_img, M, (w_, h_))

cv2.imshow(" new load", rot)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
