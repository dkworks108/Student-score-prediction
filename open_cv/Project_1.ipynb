```python
import cv2
image_loc =  input(" Enter Your file name  :- ")

if image_loc is None:
    print(" Please Enter  File Location")
else:
    imag = cv2.imread(image_loc) 
    if imag is None:
        print(" Your input file is not a image. ")
    else:
        gray_img =  cv2.cvtColor(imag, cv2.COLOR_BGR2GRAY)
        cv2.imshow(" THE gray image", gray_img)
        cv2.waitKey(0)
        cv2.destroyAllWindows()
        gray_img_name = input(" Enter file name to save ")
        
        cv2.imwrite(gray_img_name, gray_img)
```
