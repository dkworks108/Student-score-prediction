```python
import pandas as pd

df = pd.read_csv('/media/bumble/DATA_DISK/dk/python/feature_Scaling/study_performance.csv')

print(df.head(2))
```

```python
df_data = df[['gender', 'test_preparation_course', 'math_score', 'reading_score', 'writing_score']]

print(df_data)
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

scaler = StandardScaler()
X_scaler = scaler.fit_transform()

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform()
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import train_test_split
import pandas as pd

standard_scaler = StandardScaler()
standard_scaled = standard_scaler.fit_transform(df_data[['math_score', 'reading_score', 'writing_score']])

print('StandardScaler output')
print(pd.DataFrame(standard_scaled, columns=['math_score', 'reading_score', 'writing_score']))
                                                                             
```

```python
minmax_scaler =MinMaxScaler()
minmax_scaled = minmax_scaler.fit_transform(df_data[['math_score', 'reading_score', 'writing_score']])

print("\n min max scaled output")

print(pd.DataFrame(minmax_scaled, columns=['math_score', 'reading_score', 'writing_score']))
```

```python
X = df[['reading_score', 'writing_score']]
Y = df[['math_score']]

X_train, X_test, Y_train, Y_test = train_test_split(X, Y , test_size=0.2, random_state=42 )

print("\nTraining data of X (X_train)")
print(X_train)

print("\nTesting data of X (X_test)")
print(X_test)

print("\nTraining data of Y (Y_train)")
print(Y_train)

print("\nTesting data of Y (Y_test)")
print(Y_test)
```

Supervised Machine learning

```python
from sklearn.linear_model import LinearRegression

X_reading = df[['reading_score']]
Y = df[['math_score']]

model = LinearRegression()
model.fit(X_reading, Y)

reading = float(input("Enter your Reading score: "))

prediction = model.predict([[reading]])

print(f"According to Reading score {reading}, predicted Math score is: {prediction[0][0]:.2f}")
```
