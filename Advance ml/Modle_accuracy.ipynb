Mean absolute Error

```python
from sklearn.metrics import mean_absolute_error 

rel_score = [60,75,90,70,100]

pre_score = [65, 70, 100, 65, 90]


mae = mean_absolute_error(rel_score,pre_score)

print(" Mean absolute Error is ")
print(mae)
```

Mean Squar Error

```python
from sklearn.metrics import mean_squared_error

rel_score = [60,90,60,70,100]

pre_score = [65, 100,60, 65, 90]

mse = mean_squared_error(rel_score,pre_score)

print(" Mean Squre Error is ")
print(mse)
```

Root Mean Square Error

```python
from sklearn.metrics import mean_squared_error  
import numpy as np   

rel_score = [60,90,70,100, 60, 35, 20]

pre_score = [65, 100, 65, 90, 50, 40 , 30]


mse = mean_squared_error(rel_score,pre_score)

rmse = np.sqrt(mse)

print(" Root Mean Square Error is ")
print(rmse)
```
