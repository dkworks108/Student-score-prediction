confusion matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score

 
acutal_data =  [1,0,1,1,1,0,0,1,0,1]

predi_data = [1,0,1,1,0,1,0,1,0,1]

cm =  confusion_matrix(acutal_data,predi_data)

print(" This is the rsult of Confution Matrix :- ")
print(cm)

print("acuracy score is ", accuracy_score(acutal_data,predi_data))
```
