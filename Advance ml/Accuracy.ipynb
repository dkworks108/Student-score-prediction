Accuracy Score

```python
from sklearn.metrics  import accuracy_score

acutal_data =  [1,0,1,1,1,0,0,1,0,1]

predi_data = [1,0,1,1,0,1,0,1,0,1]

print("Accuracy Score is ", accuracy_score(acutal_data,predi_data))
```

Precision Score

```python
from sklearn.metrics import precision_score

acutal_data =  [1,0,1,1,1,0,0,1,0,1]

predi_data = [1,0,1,1,0,1,0,1,0,1]

print(" Precision Score is ", precision_score(acutal_data,predi_data))
```

Recall Score

```python
from sklearn.metrics import recall_score

acutal_data =  [1,0,1,1,1,0,0,1,0,1]

predi_data = [1,0,1,1,0,1,0,1,0,1]

print("Recall Score is ", recall_score(acutal_data,predi_data))
```

F1 Score

```python
from sklearn.metrics import f1_score

acutal_data =  [1,0,1,1,1,0,0,1,0,1]

predi_data = [1,0,1,1,0,1,0,1,0,1]

print("F1 Score is ", f1_score(acutal_data,predi_data))
```
