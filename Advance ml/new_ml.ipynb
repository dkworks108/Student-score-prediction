```python
import pandas as pd
import numpy as np   
import matplotlib.pyplot as plt    

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
data = {
    'Age' : [25,30,35,40,45,50],
    'Income' : [30000, 40000, 50000, 60000, 70000, 80000],
    'Spending' : [70,60,50,40,20,10],
    'Savings' : [1000, 5000, 8000, 10000, 15000,  20000]
}
```

```python
df =  pd.DataFrame(data)
```

```python
scaler =  StandardScaler()
Scaled_data = scaler.fit_transform(df)
```

```python
print(Scaled_data)
```

```python
print(df)
```

```python
pca = PCA(n_components=2)
pca_result = pca.fit_transform(Scaled_data)

pca_df = pd.DataFrame(pca_result, columns=['PCA1', 'PCA2'])
```

```python
explained_variance = pca.explained_variance_ratio_

print("Variance Components")
print(np.round(explained_variance *100, 2))
```

```python
plt.Figure(figsize=(8,6))
plt.scatter(pca_df['PCA1'], pca_df['PCA2'], color='black')
plt.title(" PCA 2d View")
plt.xlabel(" PCA 1 main Patten")
plt.ylabel(" PCA 2 main Patten")
plt.grid(True)
plt.show()

print(pca_df)
```
