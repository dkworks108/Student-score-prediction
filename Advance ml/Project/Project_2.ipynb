```python
import pandas as pd    
import numpy as np 
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score 
```

```python
stu_data = pd.read_csv('/media/bumble/DATA_DISK/dk/python/Advance ml/Project/student_performance.csv')

X = stu_data[['weekly_self_study_hours','attendance_percentage','class_participation']]
 
y = stu_data['total_score']
```

```python
model =  LinearRegression()
model.fit(X,y)

model_pre = model.predict(X)
```

```python
mae = mean_squared_error(y,model_pre)
mse = mean_squared_error(y,model_pre)
rmse = np.sqrt(mse)
r2 = r2_score(y,model_pre)
```

```python
print(f" Mean Absolute Error is {mae:.2f}")

print(f" Mean Square Error is {mse:.2f}")

print(f" Root Mean Square Error is {rmse:.2f}")

print(f" R2 score is {r2:.2f}")
```

```python
new_weekly_self_study_hours = float(input("Enter weekly self-study hours: "))
new_attendance_percentage = float(input("Enter attendance percentage: "))
new_class_participation = float(input("Enter class participation score: "))

new_prect_score = model.predict(np.array([[new_weekly_self_study_hours,new_attendance_percentage,new_class_participation]]))

print(f" Based on inputs like weely study houres {new_weekly_self_study_hours:.2f} , attendance percentage {new_attendance_percentage:.2f} and class participation is {new_class_participation:.2f}. The predicted Score is {new_prect_score[0]:.2f}")
```
