Predicting Student Score based on thire study houres.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error, mean_squared_error
import pandas as pd    
import numpy as np   
```

```python
data = pd.read_csv('/media/bumble/DATA_DISK/dk/python/Advance ml/Project/data_1.csv')


x = data[['Hours']]
y = data['Score']
```

```python
model = LinearRegression()
model.fit(x,y)
#new_hour = float(input(" Enter your study houres"))#

input_hour = 2
y_pred_score =  model.predict(np.array([[input_hour]]))


#print(f" Your score will be {pred_score[0]:.2f} on the bases of {new_hour} hours. ")#
```

```python
y_pred = model.predict(x)

print(
    f"Model accuracy using R2 score (based on actual y and predicted y values) is "
    f"{r2_score(y, y_pred):.2f}."
)
```

```python
mar = mean_absolute_error(y,y_pred)
msr = mean_squared_error(y,y_pred)
rmsr = np.sqrt(msr)
```

```python
print(f" Mean Absolute Error is {mar:.2f}")

print(f" Mean Square Error is {msr:.2f}")

print(f" Root Mean Square Error is {rmsr:.2f}")
```
