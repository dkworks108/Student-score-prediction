```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd

df = pd.read_csv('/media/bumble/DATA_DISK/dk/python/learing/data.csv')

df_label = df.copy()

le = LabelEncoder()

# Gender Encoding
df_label['Gender_Encoded'] = le.fit_transform(df_label['Gender'])

# Passed Encoding
df_label['Passed_Encoded'] = le.fit_transform(df_label['Passed'])

# Name Encoding
df_label['Name_Encoding'] = le.fit_transform(df_label['Name'])

print('\nLabel Encoded Data')
print(df_label[['Name', 'Name_Encoding','Gender', 'Gender_Encoded', 'Passed', 'Passed_Encoded']])


df_encoded = pd.get_dummies(df_label, columns=['City'])
print('\n Encoded DATA')
print(df_encoded)
```
